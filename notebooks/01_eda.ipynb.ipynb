```python
# Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings for display
pd.set_option('display.float_format', lambda x: '%.3f' % x)
sns.set(style="whitegrid")

# Data loading
df = pd.read_csv("data/creditcard.csv")

# Preview
print("Dataset shape:", df.shape)
df.head()
```

```python
# Data types, number of non-null values, memory
df.info()
```

```python
# Check for missing values
missing = df.isnull().sum()
missing[missing > 0]
```

```python
# Statistics for all numeric columns
df.describe().T
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicates: {duplicates}")
```

```python
# Delete duplicates
df = df.drop_duplicates()

# check
print("Remaining duplicates:", df.duplicated().sum())
print("New dataset shape:", df.shape)

# Cleaned data after EDA
df.to_csv("data/creditcard_cleaned.csv", index=False)
```

```python
# Class distribution (absolute and relative)
class_counts = df["Class"].value_counts()
class_percent = df["Class"].value_counts(normalize=True) * 100

# Output
print("Transaction count per class:")
print(class_counts)
print("\nPercentage per class:")
print(class_percent.round(4))
```

```python
# Class distribution
labels = ["Legitimate", "Fraud"]
counts = df["Class"].value_counts()
percentages = df["Class"].value_counts(normalize=True) * 100
colors = ["#66b3ff", "#ff6666"]

# Labels with absolute numbers and percentage
custom_labels = [
    f"{label}: {count:,} ({percentage:.4f}%)"
    for label, count, percentage in zip(labels, counts, percentages)
]

# Plot
plt.figure(figsize=(6, 6))
wedges, texts, autotexts = plt.pie(
    counts,
    labels=custom_labels,
    colors=colors,
    autopct="",
    startangle=90,
    counterclock=False,
    wedgeprops={"edgecolor": "black", "linewidth": 0.3}
)

# Styling
plt.title("Class Distribution: Fraud vs Legitimate", fontsize=14)
plt.axis("equal")

# Legend for clarity (optional)
plt.legend(wedges, labels, title="Class", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Amount"], bins=100, kde=True, color="#4682B4")
plt.title("Distribution of Transaction Amounts", fontsize=14)
plt.xlabel("Amount ($)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df["Class"] == 0]["Amount"], label="Legitimate", fill=True, color="#66b3ff")
sns.kdeplot(data=df[df["Class"] == 1]["Amount"], label="Fraud", fill=True, color="#ff6666")
plt.title("Transaction Amount by Class", fontsize=14)
plt.xlabel("Amount ($)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Add small constant to avoid log(0)
df["Log_Amount"] = np.log1p(df["Amount"])

plt.figure(figsize=(10, 6))
sns.histplot(df["Log_Amount"], bins=100, kde=True, color="#6A5ACD")
plt.title("Distribution of Log-Transformed Amount", fontsize=14)
plt.xlabel("log(Amount + 1)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df["Class"] == 0]["Time"], label="Legitimate", fill=True, color="#66b3ff")
sns.kdeplot(data=df[df["Class"] == 1]["Time"], label="Fraud", fill=True, color="#ff6666")
plt.title("Time Distribution by Transaction Type", fontsize=14)
plt.xlabel("Time (seconds)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate IQR bounds
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Extract outliers
outliers = df[(df['Amount'] < lower_bound) | (df['Amount'] > upper_bound)]

# Save to CSV file
outliers.to_csv("data/outliers_amount.csv", index=False)

print(f"Number of outliers: {outliers.shape[0]}")
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(x='Class', y='Amount', data=outliers, jitter=True, hue='Class', palette="Set1", alpha=0.5, dodge=False, legend=False)
plt.title("Outliers by Class (0 = Legitimate, 1 = Fraud)")
plt.xlabel("Class")
plt.ylabel("Amount ($)")
plt.show()
```

```python
# Select only V1–V28 + Amount
features = df.loc[:, 'V1':'V28'].copy()
features["Amount"] = df["Amount"]

# Compute correlation matrix
corr_matrix = features.corr()

# Plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of PCA Components and Amount")
plt.show()
```

```python
# Group by Class
mean_amount = df.groupby('Class')['Amount'].mean().reset_index()

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(x='Class', y='Amount', data=mean_amount, palette='Set1')
plt.title("Average Transaction Amount by Class")
plt.xlabel("Class (0 = Legitimate, 1 = Fraud)")
plt.ylabel("Average Amount ($)")
plt.show()

# Print values
print(mean_amount)
```

```python
# Create a copy with new 'Hour' column
df_time = df.copy()
df_time['Hour'] = df_time['Time'] // 3600  # Convert seconds to hours

# Count frauds per hour
fraud_per_hour = df_time[df_time['Class'] == 1].groupby('Hour').size()

# Plot
plt.figure(figsize=(12, 6))
fraud_per_hour.plot(kind='bar', color='red')
plt.title("Number of Fraudulent Transactions per Hour")
plt.xlabel("Hour since first transaction")
plt.ylabel("Fraud Count")
plt.tight_layout()
plt.show()
```
